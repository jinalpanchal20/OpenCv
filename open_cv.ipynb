# IMAGE ADD

```python
img= cv.imread('im1.jpg',0)
img=img+255
cv.imshow('j',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

# IMAGE BGR TO GRAY

```python
img= cv.imread('im1.jpg',1)
img1=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow('j',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

# IMAGE RGB TO BGR

```python
img= cv.imread('im1.jpg',1)
img1=cv.cvtColor(img,cv.COLOR_RGB2BGR)
cv.imshow('j',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img1)
```

# IMAGE TRANSLATION

```python
img= cv.imread('im1.jpg',1)
mat=np.float32([[1,0,1],
                [0,1,1]])
img1=cv.warpAffine(img,mat,(200,200))
cv.imshow('j',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img1)
```

```python
img= cv.imread('im1.jpg',1)
mat=np.float32([[1,0,20],
                [0,1,10]])
img1=cv.warpAffine(img,mat,(300,300))
cv.imshow('j',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

# IMAGE ROTATION

```python
import math as m
```

```python
img= cv.imread('im1.jpg',1)
mat=np.float32([[m.cos(45),m.sin(45),0],[-m.sin(30),m.cos(30),0]])
img1=cv.warpAffine(img,mat,(500,500))
cv.imshow('j',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img1)
```

# IMAGE SHARING

```python
img= cv.imread('im1.jpg',1)
c,r,b=img.shape
mat=np.float32([[1,2,1],
                [2,1,1]])
img1=cv.warpAffine(img,mat,(c+300,r+300))
cv.imshow('j',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img1)
```

```python
img= cv.imread('img.jpg',1)
i=cv.resize(img,(200,200))
t=cv.getRotationMatrix2D((100,100),45,1)
c,r,b=img.shape
img1=cv.warpAffine(i,t,(c,r))
cv.imshow('j',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```
