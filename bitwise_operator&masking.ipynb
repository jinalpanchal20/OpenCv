```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
img = cv.imread("shivji.jpeg")
img1 = cv.resize(img,(500,500))
cv.imshow("d1",img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img2 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
img3 = cv.resize(img2,(500,500))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img1)
plt.title("original image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img3)
plt.title("rgb image")
plt.axis("off")
plt.show()
```

```python
i1 = cv.imread("shivji.jpeg")
i2 = cv.imread("ambe.jpeg")
i3=cv.resize(i1,(500,500))
i4=cv.resize(i2,(500,500))
i5 = cv.add(i3,i4)
plt.imshow(i5)
plt.show()
```

```python
i6 = cv.subtract(i3,i4)
plt.imshow(i6)
```

```python
i7= cv.subtract(i3,i3)
plt.imshow(i7)
```

```python
constant = 4
i8 = cv.multiply(i3,constant)
plt.imshow(i8)
```

```python
constant = 2.2
i9 = cv.divide(i3,constant)
plt.imshow(i9)
```

```python
i10 = cv.bitwise_and(i3,i4)
plt.imshow(i10)
```

```python
i11 = cv.bitwise_or(i3,i4)
plt.imshow(i11)
```

```python
i12 = cv.bitwise_not(i3,i4)
plt.imshow(i12)
```

```python
i13 = cv.addWeighted(i3,0.6,i4,0.4,0)
cv.imshow("d3",i13)
cv.waitKey(0)
cv.destroyAllWindows()
```
