```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = cv.imread("shivji.jpeg")
i1 = cv.cvtColor(i,cv.COLOR_BGR2RGB)
```

```python
def four_neighbor(i1,x,y):
    print("4 neighbors with pixel values")
    print("(",x+1,",",y,") = ",i1[x+1,y])
    print("(",x-1,",",y,") = ",i1[x-1,y])
    print("(",x,",",y+1,") = ",i1[x,y+1])
    print("(",x,",",y-1,") = ",i1[x,y-1])
    
```

```python
four_neighbor(i1,10,20)
```

```python
def distance(p1,p2):
    d = 0
    for i in range(len(p1)):
        d += (p1[i]-p2[i])**2
    print("distance = ",np.round(np.sqrt(d),4))
p1 = (2,3)
p2 = (6,7)
distance(p1,p2)
```

```python
def dist(p3,p4):
    d1 = 0
    for i in range(len(p3)):
        d1 += (p3[i]-p4[i])**2
    print("the euclidean distance is = ",np.round(np.sqrt(d1),3))
p3 = (2,3)
p4 = (6,7)
```

```python
dist(p3,p4)
```
