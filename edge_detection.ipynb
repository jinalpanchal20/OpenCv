```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("d1.jpeg")
img = cv.cvtColor(img,cv.COLOR_RGB2BGR)
```

```python
def rober(img):
    k1 = np.array([[1,0],[0,-1]])
    k2 = np.array([[0,1],[-1,0]])
    rx = cv.filter2D(img,-1,k1)
    ry = cv.filter2D(img,-1,k2)
    rober1 = np.sqrt(np.square(rx)+np.square(ry))
    rober1 = rober1.astype('int32')
    return rober1
```

```python
img_rober = rober(img)
```

```python
def sobel(img):
    k1 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    k2 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
    sx = cv.filter2D(img,-1,k1)
    sy = cv.filter2D(img,-1,k2)
    sobel1 = np.sqrt(np.square(sx)+np.square(sy))
    sobel1 = sobel1.astype('int32')
    return sobel1
```

```python
img_sobel = sobel(img)
```

```python
def prewitt(img):
    k1 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    k2 = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
    px = cv.filter2D(img,-1,k1)
    py = cv.filter2D(img,-1,k2)
    prewitt1 = np.sqrt(np.square(px)+np.square(py))
    prewitt1 = prewitt1.astype('int32')
    return prewitt1
```

```python
img_prewitt = prewitt(img)
```

```python
img_canny = cv.Canny(img,50,150)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(3, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original Image')
plt.subplot(3, 2, 2), plt.imshow(img_rober, cmap='gray'), plt.title('Robber Edge')
plt.subplot(3, 2, 3), plt.imshow(img_sobel, cmap='gray'), plt.title('Sobel Edge')
plt.subplot(3, 2, 4), plt.imshow(img_prewitt, cmap='gray'), plt.title('Perwitt Edge')
plt.subplot(3, 2, 5), plt.imshow(img_canny, cmap='gray'), plt.title('Canny Edge')
plt.tight_layout()
plt.show()
```
